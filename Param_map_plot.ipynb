# Define the Function
Adapt the 'plot_parameter_maps' function to take three lists of paths to images and plot them in a matrix.

```python
# Define the Function
def plot_image_matrix(list1, list2, list3, row_labels, col_labels, save_path=None):
    """
    Plots images from three lists in a matrix, each list gets its own row and every entry is a column.
    
    Parameters:
    - list1, list2, list3 (list of str): Lists of paths to the images.
    - row_labels (list of str): Labels for the rows.
    - col_labels (list of str): Labels for the columns.
    - save_path (str, optional): Path to save the resulting plot.
    """
    # Combine the lists into a single list of lists
    image_lists = [list1, list2, list3]
    
    # Create the figure with tighter grid spacing
    plt.style.use("dark_background")
    fig, axes = plt.subplots(len(image_lists), len(list1), figsize=(2.5 * len(list1), 2.5 * len(image_lists)),
                             gridspec_kw={"wspace": 0.02, "hspace": 0.02})  # Tighter spacing

    font_size = 14

    for row_idx, image_list in enumerate(image_lists):
        for col_idx, image_path in enumerate(image_list):
            # Load the image
            img = nib.load(image_path).get_fdata()
            
            # Plot the image
            ax = axes[row_idx, col_idx]
            ax.imshow(img[:, :, img.shape[2] // 2], cmap='gray')  # Display the middle slice
            ax.axis('off')
            
            # Add row labels
            if col_idx == 0:
                ax.set_ylabel(row_labels[row_idx], fontsize=font_size)
            
            # Add column labels
            if row_idx == 0:
                ax.set_title(col_labels[col_idx], fontsize=font_size)

    # Save the plot if save_path is provided
    if save_path:
        plt.savefig(save_path, bbox_inches='tight')
    
    plt.show()
```

# Create Sample Data
Create sample data with three lists of paths to images for testing the function.

```python
# Create Sample Data

# Sample lists of paths to images
list1 = ["path/to/image1_1.nii", "path/to/image1_2.nii", "path/to/image1_3.nii"]
list2 = ["path/to/image2_1.nii", "path/to/image2_2.nii", "path/to/image2_3.nii"]
list3 = ["path/to/image3_1.nii", "path/to/image3_2.nii", "path/to/image3_3.nii"]

# Row and column labels
row_labels = ["Row 1", "Row 2", "Row 3"]
col_labels = ["Col 1", "Col 2", "Col 3"]

# Call the function to plot the images
plot_image_matrix(list1, list2, list3, row_labels, col_labels)
```
